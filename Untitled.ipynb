# Sentiment analysis 

Perform sentiment analysis on video comments to measure viewer perceptions. This task involves basic processing of text data and visualizing sentiment trends across various video categories.

```python
#  sentiment preprocessing
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()



def preprocess_text(text):
  tokens = word_tokenize(text.lower())
  tokens = [word for word in tokens if word.isalnum()]
  tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
  return ' '.join(tokens)


youtube["preprocessed_comments"] = youtube["Comment"].apply(preprocess_text)
```

```python
# I do not want to rely on “sentiment analysis score” in analysis. I want to make sure and come up with a new “sentiment score” to rely on

def get_sentiment_score(text):
    analysi = TextBlob(text)
    polarity = analysi.sentiment.polarity
    if polarity > 0:
        return 2
    elif polarity < 0:
        return 0
    else:
        return 1

youtube["sentiment_score_2"] = youtube["preprocessed_comments"].apply(get_sentiment_score)
```

```python
# santiment analysis value counts
postive = youtube[youtube["sentiment_score_2"] == 2]
negative = youtube[youtube["sentiment_score_2"] == 0]
neutral = youtube[youtube["sentiment_score_2"] == 1]
print("Positive:",len(postive))
print("Negative:",len(negative))
print("Neutral:",len(neutral))
```

```python
youtube["sentiment_score_2"].value_counts().plot(kind="bar")
plt.title("Sentiment Analysis")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.xticks([0, 1, 2], ["Positive", "Neutral", "Negative"])
plt.show()
```

```python
# visualize sentiment
sentiment_trends = youtube.groupby(['Keyword', 'sentiment_score_2']).size().unstack().head(10)
sentiment_trends.plot(kind='bar', stacked=True,colormap='viridis')
plt.title('Sentiment Trends by Keyword')
plt.xlabel('Keyword')
plt.ylabel('Count')
plt.show()
```

# Ranking model

Development of a Video Ranking Model: Create a simple model that uses sentiment analysis results and traditional engagement metrics to rank videos. This model should help identify potentially valuable videos for specific industry sectors.

```python
rank_model= youtube.copy()
```

```python
rank_model.head(1)
```

```python
# Scaler Dataset 
scaler= MinMaxScaler()
rank_model[['Video_Likes', 'Video_Comments', 'Video_Views','comment_like']] = scaler.fit_transform(
    rank_model[['Video_Likes', 'Video_Comments', 'Video_Views','comment_like']])
```

```python
# calculate Engagement score 
rank_model['Engagement_Score'] = rank_model[['Video_Likes', 'Video_Comments', 'Video_Views','comment_like']].sum(axis=1)
```

```python
# calculate Rank featuer 
rank_model['Rank'] = rank_model['Engagement_Score'].rank(ascending=False)
```

```python
rank_model.head()
```

```python
# Split data featuers & target var 
features = rank_model[['Video_Likes', 'Video_Comments', 'Video_Views', 'comment_like']]
target = rank_model['Rank']
```

```python
# Split train & testset 
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('model', RandomForestRegressor(random_state=42))
])
```

```python
param_grid = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(f'Best Parameters: {grid_search.best_params_}')
print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
```
